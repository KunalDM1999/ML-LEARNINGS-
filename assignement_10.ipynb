```python
import pandas as pd
path ='/content/used_cars.csv'
df = pd.read_csv(path)
df
```

```python
df.info()
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingRegressor, RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
x=df.drop('MSRP',axis=1)
y=df['MSRP']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
# Initialize the models
bagging_model = BaggingRegressor(random_state=1)
rf_model = RandomForestRegressor(random_state=1)
gb_model = GradientBoostingRegressor(random_state=1)
ada_model = AdaBoostRegressor(random_state=1)

# Train the models
bagging_model.fit(x_train, y_train)
rf_model.fit(x_train, y_train)
gb_model.fit(x_train, y_train)
ada_model.fit(x_train, y_train)
print('Score on Testing set is BaggingRegressor:',bagging_model.score(x_test,y_test))
print('Score on Testing set is RandomForestRegresso:',rf_model.score(x_test,y_test))
print('Score on Testing set is GradientBoostingRegressor:',gb_model.score(x_test,y_test))
print('Score on Testing set is AdaBoostRegressor:',ada_model.score(x_test,y_test))
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import AdaBoostRegressor

param_grid = {
    'n_estimators': [100, 500, 1000],
    'learning_rate': [0.5, 1, 2]
}

ada_model = AdaBoostRegressor(random_state=1)
grid_search = GridSearchCV(estimator=ada_model, param_grid=param_grid, cv=4)
grid_search.fit(x_train, y_train)

# Best model and score
best_model = grid_search.best_estimator_
score = best_model.score(x_test, y_test)

print("Best Hyperparameters:", grid_search.best_params_)
print('Score on Testing set is:', score)
```
