```python
import pandas as pd

# Load the dataset
flights_df = pd.read_parquet('/content/flights.parquet')
# Convert scheduled departure to hours for easier comparison

# Filter flights traveling less than 250 miles and with scheduled departures between 14:00 and 16:00
filtered_df = flights_df[(flights_df['DISTANCE'] < 250) &
                         (flights_df['SCHEDULED_DEPARTURE'] >= 14) &
                         (flights_df['SCHEDULED_DEPARTURE'] < 16)]
# Drop rows with NaN values in DEPARTURE_DELAY or ARRIVAL_DELAY
filtered_df = filtered_df.dropna(subset=['DEPARTURE_DELAY', 'ARRIVAL_DELAY'])

# Calculate Pearson correlation
correlation = filtered_df['DEPARTURE_DELAY'].corr(filtered_df['ARRIVAL_DELAY'])
print(f'The Pearson correlation between DEPARTURE_DELAY and ARRIVAL_DELAY is: {correlation}')
```

```python
import pandas as pd

# Load the dataset
file_path = '/content/flights.parquet'
df = pd.read_parquet(file_path)

# Filter flights traveling less than 1950 miles
df_filtered = df[df['DISTANCE'] < 1950]

# Further filter to include only flights with scheduled departure times between 20:00 and before 21:00
df_filtered = df_filtered[(df_filtered['SCHEDULED_DEPARTURE'] >= 20.0) & (df_filtered['SCHEDULED_DEPARTURE'] < 21.0)]

# Remove rows where either DEPARTURE_DELAY or ARRIVAL_DELAY is blank
df_filtered.dropna(subset=['DEPARTURE_DELAY', 'ARRIVAL_DELAY'], how='any', inplace=True)
# Calculate IQR of arrival delays
Q1 = df_filtered['ARRIVAL_DELAY'].quantile(0.25)
Q3 = df_filtered['ARRIVAL_DELAY'].quantile(0.75)
IQR = Q3 - Q1

# Calculate lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
# Count number of outliers
outliers = df_filtered[(df_filtered['ARRIVAL_DELAY'] < lower_bound) | (df_filtered['ARRIVAL_DELAY'] > upper_bound)]
num_outliers = outliers.shape[0]

num_outliers
```

```python
import pandas as pd

# Load the dataset
file_path = '/content/flights.parquet'
df = pd.read_parquet(file_path)

# Filter flights traveling less than 1650 miles
df_filtered = df[df['DISTANCE'] < 1650]

# Further filter to include only flights with scheduled departure times between 16:00 and before 17:00
df_filtered = df_filtered[(df_filtered['SCHEDULED_DEPARTURE'] >= 16.0) & (df_filtered['SCHEDULED_DEPARTURE'] < 17.0)]

# Remove rows where either DEPARTURE_DELAY or ARRIVAL_DELAY is blank
df_filtered.dropna(subset=['DEPARTURE_DELAY', 'ARRIVAL_DELAY'], how='any', inplace=True)
# Calculate IQR of arrival delays
Q1 = df_filtered['ARRIVAL_DELAY'].quantile(0.25)
Q3 = df_filtered['ARRIVAL_DELAY'].quantile(0.75)
IQR = Q3 - Q1

# Calculate lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
# Count number of outliers
outliers = df_filtered[(df_filtered['ARRIVAL_DELAY'] < lower_bound) | (df_filtered['ARRIVAL_DELAY'] > upper_bound)]
num_outliers = outliers.shape[0]

num_outliers
```

```python
import pandas as pd

# Load the dataset
file_path = '/content/house-rent.csv'
df = pd.read_csv(file_path)
from sklearn.linear_model import LinearRegression
import numpy as np

# Filter dataset for Unfurnished houses in Chennai
chennai_unfurnished = df[(df['City'] == 'Chennai') & (df['Furnishing Status'] == 'Unfurnished')]

# Filter dataset for Unfurnished houses in Hyderabad
hyderabad_unfurnished = df[(df['City'] == 'Hyderabad') & (df['Furnishing Status'] == 'Unfurnished')]

# Define a function to calculate the regression slope
def calculate_regression_slope(dataframe):
    X = dataframe['Size'].values.reshape(-1, 1)
    y = dataframe['Rent'].values
    model = LinearRegression()
    model.fit(X, y)
    slope = model.coef_[0]  # Get the slope (coefficient) of the regression line
    return slope

# Calculate regression slopes
slope_chennai = calculate_regression_slope(chennai_unfurnished)
slope_hyderabad = calculate_regression_slope(hyderabad_unfurnished)

# Calculate the absolute difference between the slopes
slope_difference = abs(slope_chennai - slope_hyderabad)

slope_chennai, slope_hyderabad, slope_difference
```

```python

# Filter dataset for Unfurnished houses in Bangalore
bangalore_unfurnished = df[(df['City'] == 'Bangalore') & (df['Furnishing Status'] == 'Unfurnished')]

# Calculate regression slopes
slope_bangalore = calculate_regression_slope(bangalore_unfurnished)

# Calculate the absolute difference between the slopes for Chennai and Bangalore
slope_difference_chennai_bangalore = abs(slope_chennai - slope_bangalore)

slope_chennai, slope_bangalore, slope_difference_chennai_bangalore
```
