```python
# prompt: lode data using url

import pandas as pd
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00488/Live_20210128.csv'
df = pd.read_csv(url)

df.info()
df.shape
```

```python
df.drop(columns=df.columns[12:], inplace=True)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

# Save 'status_type' as target variable
y = df['status_type']

# Drop the feature "status_id", "status_type" and "status_published" from training set
X = df.drop(["status_id", "status_type", "status_published"], axis=1)

# Use LabelEncoder to transform the target variable
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

# Use standard scaler to scale the features
scaler = StandardScaler()
X = scaler.fit_transform(X)
df['status_type'].nunique()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=10)
kmeans.fit(X)
print(kmeans.inertia_)
```

```python
kmeans2 = KMeans(n_clusters=4, random_state=10)
kmeans2.fit(X)

labels = kmeans2.labels_
correct_labels = sum(y == labels)
print("Result: %d out of %d samples were correctly labeled." % (correct_labels, y.size))
```

```python
from sklearn.cluster import AgglomerativeClustering

clustering_model = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
clustering_model.fit(X)
print("Predicted label for the first row:", clustering_model.labels_[0])
```

```python
n_samples = X.shape[0]
n_leaves = clustering_model.n_leaves_
print("Number of leaves in the hierarchical tree:", n_leaves)
```

```python
labels = clustering_model.labels_
correct_labels = sum(y == labels)
print("Result: %d out of %d samples were correctly labeled." % (correct_labels, y.size))
print('Accuracy score: {0:0.2f}'. format(correct_labels/float(y.size)))
```

```python

import pandas as pd
url = '/content/AirQualityUCI.csv'
df = pd.read_csv(url)

df.info()
df.shape
```

```python
df.drop(columns=df.columns[15:], inplace=True)
df.drop(columns=df.columns[:2], inplace=True)
df.dropna(inplace=True)
df
```

```python
df['T'].corr(df['RH'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor
from sklearn.pipeline import make_pipeline

X = df.drop('AH', axis=1)
y = df['AH']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

pipeline = make_pipeline(
    StandardScaler(),
    MLPRegressor(hidden_layer_sizes=(50, 50, 50),
                 random_state=1,
                 tol=1e-2,
                 alpha=1e-4,
                 solver="adam",
                 learning_rate_init=0.1,
                 max_iter=1)
)
X_test.shape[0]
```

```python
pipeline.fit(X_train, y_train)
y_train_pred = pipeline.predict(X_train)
y_test_pred = pipeline.predict(X_test)
# Calculate R^2 (coefficient of determination) on training set
from sklearn.metrics import r2_score
r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test,y_test_pred)
print("R^2 on training set: {:.2f}".format(r2_train))
print("R^2 on test set: {:.2f}".format(r2_test))
print("Prediction for the first instance of the training set:", pipeline.predict(X_train.iloc[[0]]))
```
